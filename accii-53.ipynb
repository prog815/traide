```python
import yfinance as yf
import pandas as pd
import os

def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
```

# Список акций

```python
acc_list = "TGKA	FEES	VTBR	TGKB	HYDR	TGKD	TGKN	RSTI	ENRU	SNGS	MRKV	GAZP	MRKZ	MRKP	SNGSP	MAGN	TGKDP	MSNG	MRKC	MRKU	NMTP	NLMK	MTLR	AFLT	MSRS	ROSN	TTLK	FESH	TRMK	RSTIP	RTKM	SIBN	AMEZ	ISKJ	DVEC	RASP	RTKMP	UTAR	LSNG	BSPB	MRKS	MVID	NKNCP	LSNGP	PIKK	KMAZ	IRKT	SVAV	CHEP	MRKK	KUBE	ROSB	KROT".split('\t')
acc_list = [f'{name}.ME' for name in acc_list]
print(acc_list)
```

# Предыдущие значения

```python
prev_close_data_file_path = find('close_data.pkl','..')
prev_close_data_file_path
```

```python
last_close_data = pd.read_pickle(prev_close_data_file_path)
last_close_data
```

# Загрузка истории курсов акций

```python
all_data = yf.download(acc_list,period='10y',threads=False)
all_data.index = pd.to_datetime(all_data.index,format="%Y-%m-%d")
all_data
```

# Выгружаем на диск полную версию выгрузки

```python
all_data.to_csv('all_data.csv')
all_data.to_pickle('all_data.pkl')
```

# Выделяем курсы закрытия

```python
close_data = all_data.Close
close_data
```

# Заполняем пустоты из предыдущей версии

```python
for acc in list(close_data.columns):
    c = close_data[acc]
    lc = last_close_data[acc]
    
    c_index = c.index[c.isna()]
    ints_index = lc.index.intersection(c_index)
#     if(len(ints_index)>0):
    c[ints_index] = lc[ints_index].values
        
close_data
```

# Выгружаем курсы закрытия на диск (выход тетрадки)

```python
close_data.to_csv('close_data.csv')
close_data.to_pickle('close_data.pkl')
```
